```python
import pandas as pd
```

```python
diabetes = pd.read_csv("./diabetes_data_upload.csv")
```

```python
diabetes
```

```python
diabetes.head()
```

```python
diabetes_gen=diabetes.groupby(['Gender']).count()
diabetes_gen = diabetes_gen.rename(columns={'Age':'count'})
diabetes_gen 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(diabetes_gen.index, diabetes_gen['count'])
plt.title('Conteo por edad')
plt.show()
```

```python
diabetes_class=diabetes.groupby(['class']).count()
diabetes_class = diabetes_class.rename(columns={'Age':'count'})
diabetes_class 
```

```python
plt.bar(diabetes_class.index, diabetes_class['count'])
plt.title('')
plt.show()
```
